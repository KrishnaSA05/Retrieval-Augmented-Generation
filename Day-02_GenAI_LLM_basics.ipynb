```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True) # Load environment variables from .env file
openai_api_key = os.getenv('OPENAI_API_KEY') # Get the OpenAI API key from environment variables
```

```python
# Connect to OpenAI
openai = OpenAI() 
```

```python
system_message = "You are an assistant that is great at telling jokes" # System prompt for the LLM
user_prompt = "Tell a light-hearted joke for an audience of Machine Learning" # User prompt for the LLM
```

```python
prompts = [ # List of messages for the chat model
    {"role": "system", "content": system_message}, 
    {"role": "user", "content": user_prompt}
  ]
```

```python
# This line sends your prompts to the OpenAI API and asks the model to generate a response
completion = openai.chat.completions.create(model='gpt-3.5-turbo', messages=prompts) #we can also here temperature, top_p, n, etc.
# Print the generated response from the model
print(completion.choices[0].message.content)
```
